dodaj do ramki danych kolumnę z wyliczoną średnią ilością czasu wolnego, którym student dysponuje w ciągu dnia.


### Wyliczając wartość czasu wolnego przyjmij następujące założenia:
- czas wolny studenta to cały czas, którego **NIE** poświęca na naukę, ćwiczenia, sen oraz siedzenie przed ekranem oraz na jedzenie i czynności higieniczne

### Następnie wyświetl otrzymaną ramkę danych tak, by spełniała następujące kryteria:
- wyświetl wyłącznie dodatkową kolumnę oraz kolumny wykorzystane do wyliczenia ilości czasu wolnego
- kolumna z czasem wolnym powinna wyświetlać się jako pierwsza z lewej

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "student_lifestyle_performance_dataset.csv"

df2 = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "sehaj1104/student-lifestyle-and-academic-performance-dataset",
  file_path,
)
print()

organised_time = df2['Study_Hours_per_Day'] + df2['Sleep_Hours'] + df2['Screen_Time_Hours'] + (df2['Gym_Hours_per_Week'] / 7) - 1 + 2.5
df2['Free_time'] = round(24 - organised_time, 2)
df2['Eating_and_hygiene'] = 2.5

wanted_columns_df2 = df2[['Free_time', 'Study_Hours_per_Day', 'Sleep_Hours', 'Screen_Time_Hours', 'Gym_Hours_per_Week', 'Eating_and_hygiene']]
print(wanted_columns_df2)
```

- Jaka dziedzina naukowa (*Branch*) jest najbardziej stresująca?
- Czy mediana snu wśród studentów informatyki (*IT*) jest wyższa niż wśród studentów mechaniki (*Mechanical*)?

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "student_lifestyle_performance_dataset.csv"

df3 = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "sehaj1104/student-lifestyle-and-academic-performance-dataset",
  file_path,
)
print()

df3 = df3[df3['Age'] >= 18]
stats = df3.groupby('Branch').agg({
    'Stress_Level_1_to_10': ['mean', 'max', 'min'],
    'Sleep_Hours': ['median']
})
stats = stats.sort_values(by=('Stress_Level_1_to_10', 'mean'), ascending=False)
print(stats)
```

## Zadanie 4
Przedstaw na wykresach:
- zależność między czasem snu a osiąganą średnią z uwzględnieniem diety studenta
- średnią obecność na zajęciach dla poszczególnych dziedzin naukowych
### UWAGA!
potraktuj informatykę (*IT*) i inżynierię komputerową (*CSE*) jako jedną dziedzinę

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "student_lifestyle_performance_dataset.csv"

df4 = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "sehaj1104/student-lifestyle-and-academic-performance-dataset",
  file_path,
)
print()
sns.scatterplot(
    data=df4,
    x="Sleep_Hours",
    y="CGPA",
    hue="Diet_Type"
)


plt.title("Zależność między czasem snu a osiąganą średnią z uwzględnieniem diety studenta")
plt.xlabel("Czas snu (h)")
plt.ylabel("Średnia ocen")

plt.show()

df4['Branch'] = df4['Branch'].replace('CSE', 'IT')
sns.barplot(
    data=df4,
    x="Branch",
    y="Attendance_Percentage"
)


plt.title("średnią obecność na zajęciach dla poszczególnych dziedzin naukowych")
plt.xlabel("Dziedziny naukowe")
plt.ylabel("Średnia obecność (%)")
plt.show()
```

## Zadanie 5:
Utwórz wykresy, które pozwalają przedstawić następujące informacje:
- Najpopularniejsze tematy filmów na YouTube Shorts w ostatnim kwartale 2024 roku
- Jaki hasztag najbardziej opłaca się umieszczać w  postach na poszczególnych platformach?

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "social_media_viral_content_dataset.csv"

df5 = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "aliiihussain/social-media-viral-content-and-engagement-metrics",
  file_path
)
print()

df5['post_datetime'] = pd.to_datetime(df5['post_datetime'])
data_ytshorts_q4 = df5[(df5['platform'] == "YouTube Shorts") & (df5['post_datetime'] >= "2024-10-01 00:00:00") & (df5['post_datetime'] <= "2024-12-31 23:59:59")].copy()

data_ytshorts_q4['total_interactions'] = data_ytshorts_q4['likes'] + data_ytshorts_q4['comments'] + data_ytshorts_q4['shares']
topic_popularity = data_ytshorts_q4.groupby('topic')['total_interactions'].mean().reset_index()
topic_popularity = topic_popularity.sort_values(by="total_interactions", ascending=False)

sns.barplot(
    data=topic_popularity,
    x="total_interactions",
    y="topic"
)


plt.title("Najpopularniejsze tematy filmów na YouTube Shorts w ostatnim kwartale 2024 roku")
plt.xlabel("Popularność (likes + comments + shares)")
plt.ylabel("tematy")
plt.show()

```
