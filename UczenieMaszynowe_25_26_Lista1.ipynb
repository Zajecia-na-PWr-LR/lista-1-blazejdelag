### Następnie wyświetl otrzymaną ramkę danych tak, by spełniała następujące kryteria:

- kolumna z czasem wolnym powinna wyświetlać się jako pierwsza z lewej
- czas wolny powinien być zaokrąglony do drugiej liczby po przecinku

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "student_lifestyle_performance_dataset.csv"

df2 = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "sehaj1104/student-lifestyle-and-academic-performance-dataset",
  file_path,
)
print()

organised_time = df2['Study_Hours_per_Day'] + df2['Sleep_Hours'] + df2['Screen_Time_Hours'] + (df2['Gym_Hours_per_Week'] / 7) - 1 + 2.5
df2['Free_time'] = round(24 - organised_time, 2)
df2['Eating_and_hygiene'] = 2.5

wanted_columns_df2 = df2[['Free_time', 'Study_Hours_per_Day', 'Sleep_Hours', 'Screen_Time_Hours', 'Gym_Hours_per_Week', 'Eating_and_hygiene']]
print(wanted_columns_df2)
```
